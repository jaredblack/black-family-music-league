```python
import pandas as pd
import os

julia_id = 'dd77ebc3b461474daa9e9b43c4377cd5'
jared_id = '53954cca040b418fbc8f42043c323187'
base_dir = './data'
```

```python
julia_votes = pd.DataFrame()

for directory in os.listdir(base_dir):
    if os.path.isdir(os.path.join(base_dir, directory)):
        csv_path = os.path.join(base_dir, directory, 'votes.csv')
        
        if os.path.exists(csv_path):
            df = pd.read_csv(csv_path)
            filtered_df = df[df['Voter ID'] == julia_id]
            
            julia_votes = pd.concat([julia_votes, filtered_df], ignore_index=True)
```

```python
julia_votes.sort_values(by="Points Assigned", ascending=False)
```

```python
# Now, pull all the submissions from all submissions.csv files into a dataframe
submissions = pd.DataFrame()

for directory in os.listdir(base_dir):
    if os.path.isdir(os.path.join(base_dir, directory)):
        csv_path = os.path.join(base_dir, directory, 'submissions.csv')
        
        if os.path.exists(csv_path):
            df = pd.read_csv(csv_path)
            submissions = pd.concat([submissions, df], ignore_index=True)

submissions
```

```python
# My songs ranked by Julia points
julia_top_voted = julia_votes.merge(submissions, on='Spotify URI', suffixes=('', '_submissions'))[['Title', 'Artist(s)', 'Points Assigned', 'Comment']]
julia_top_voted.sort_values(by="Points Assigned", ascending=False).head(50)
```
